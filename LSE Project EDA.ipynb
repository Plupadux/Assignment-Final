```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from datetime import datetime
```

```python
ldn = pd.read_csv('/Users/rebeccaswedberg/Downloads/LSE_DA_Employer_Project_2_London/London/Central London.csv')

ldn.head()
```

```python
#Removing the first 3 letters in the Survey Date 

ldn['Survey date'] = ldn['Survey date'].str[4:]
ldn
```

```python
#Dropping Unnamed columns 
ldn = ldn.drop(columns=['Unnamed: 14', 'Unnamed: 15',"Unnamed: 16"])

ldn.head()
```

```python
# Converting the Survey date to Date time 
ldn['Survey date'] = pd.to_datetime(ldn['Survey date'])
ldn.info()
```

```python
# Checking for NaN values 

ldn.isnull().sum()
```

```python
#Removing Nan values
ldn2 = ldn.dropna(axis = 0, how ='any')
  
print(ldn2.isnull().sum())
ldn2.shape
```

```python
ldn2.info()
```

```python
#Keeping only the date in the Calendar Quarter column and changing the name 

#Renaming column

ldn2= ldn2.rename({'Survey wave (calendar quarter)':'Quarter','Number of cycle hire bikes':'Number of hired bikes' }, axis = 'columns')


ldn2['Quarter'] = ldn2['Quarter'].str[:7]

ldn2.head()
```

# Private Cycles per Month

```python
pd.pivot_table(ldn2,index=ldn2['Survey date'].dt.month,columns=ldn2['Survey date'].dt.year,
               values='Number of private cycles',aggfunc=np.sum)
```

```python
pd.pivot_table(ldn2,index=ldn2['Survey date'].dt.month,columns=ldn2['Survey date'].dt.year,
               values='Number of private cycles',aggfunc=np.sum).plot()
plt.ylabel('Private Cycles')
plt.xlabel('Month')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.title('Private Cycles per Month')
```

Exploring what revenue the Santander bike scheme is brining in, if this is something that could promp further analysis

Assuming the price for each rental is 2 Pounds, how much revenue will be generated over time and what is the predicted revenue over the next couple of years 

```python
# Grouping by Survey date and number of rentals to see how many bikes were rented per day. Creating a new dataset 

revs = ldn2[['Survey date','Time','Number of hired bikes']]

revs.head()
```

```python
revs['Revenue']= revs['Number of hired bikes'] *2

revs
```

```python
# create scatterplot with Survey date and Revenue to get an understanding of the revenue per year 
sns.scatterplot(x="Survey date", y="Revenue", data=revs)
```

```python
# Pivot table showing the revenue per month per year 

pd.pivot_table(revs,index=revs['Survey date'].dt.month,columns=revs['Survey date'].dt.year,
               values='Revenue',aggfunc=np.sum)
```

```python
revs.pivot_table(index=[revs['Survey date'].dt.month,revs['Survey date'].dt.year],
               values='Revenue',aggfunc=np.sum)
```

```python
pd.pivot_table(revs,index=revs['Survey date'].dt.month,columns=revs['Survey date'].dt.year,
               values='Revenue',aggfunc=np.sum).plot()
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.title('Revenue per Month')
```

# Weather

```python
#Checking the values in the Weather column 

ldn2['Weather'].value_counts()
```

```python
#Mapping the Weather column to Wet, Dry, Rain and Sunny. This will make it easier to analyze the data and there are
#multiple rows which are containing the same value but with a different spelling 

ldn2['Weather']= ldn2.Weather.map({'Wet': 'Wet', 'Dry':'Dry', 'Rain':'Rain','Sunny':'Sunny', 'Cloudy':'Cloudy','Fine':'Fine'})

ldn2['Weather'].value_counts()
```

```python
# replacing missing values with Other for Weather
ldn2['Weather'].fillna('Other', inplace=True)

# calculate missing values for Weather 
ldn2['Weather'].isnull().sum()
```

```python
ldn2['Weather'].value_counts()
```

```python
#Creating pivot table for what time people hire a bike 
pd.pivot_table(ldn2,index=ldn2['Weather'],values='Number of hired bikes',aggfunc=np.sum).plot(kind='bar')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.ylabel('Amount of hired bikes')
plt.xlabel('Type of Weather')
plt.title('Hired bies per Weather condition')
```
